# OpenCV test

## Method 1

```python
import cv2
from urllib.request import urlopen
from scipy.misc import imread
from matplotlib import pyplot as plt
import numpy as np
```

```python
url="https://us.123rf.com/450wm/spvvk/spvvk1612/spvvk161200093/68328971-stanford-university-main-quad.jpg"
```

```python
with urlopen(url) as file:
    img = imread(file, mode='RGB')
```

```python
type(img)
```

```python
img=np.asarray(img)
type(img)
```

```python
plt.imshow(img)
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(gray)
```

## Method 2

```python
from PIL import Image
import requests
from io import BytesIO
```

```python
response = requests.get(url)
img2 = Image.open(BytesIO(response.content))
```

```python
type(img2)
```

```python
img2 = np.asarray(img2)
```

```python
plt.imshow(img2)
```

```python
gray2 = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(gray2)
```
